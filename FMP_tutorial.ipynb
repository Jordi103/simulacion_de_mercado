Finalmente añadimos la api key: ```&apikey={NUESTRA_API_KEY}```
El resultado es el siguiente enlace:

https://financialmodelingprep.com/stable/historical-price-eod/full?symbol=AAPL&apikey={NUESTRA_API_KEY}
